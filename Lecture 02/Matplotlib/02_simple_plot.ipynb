```python


# Define the function piecewise
def x_t(t):
    return np.piecewise(t, [t < -1, (t >= -1) & (t < 0),(t >= 0) & (t < 1), t >= 1],
                           [0,
                               lambda t: 2 + 2*t,     # Line from -1 to 0 (slope 1)
                            2,     # Line from 0 to 1 (slope 1)
                            0])                  # Constant 2 for t >= 1

# Define the range of t
t = np.linspace(-2, 2, 400)

# Plot the function
plt.plot(t, (1/2)*(x_t(t)-x_t(-t)), label=r'$x(t)$', color='black')

# Add title and labels
plt.title(r'Plot of $x(t)$')
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')

# Mark specific points
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# Display grid
plt.grid(True)

# Set limits for better visualization
# plt.xlim([-2, 2])
# plt.ylim([-4, 3])

# Show plot
plt.show()
```
