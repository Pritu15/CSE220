```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values=np.linspace(0,2,100)
# print(x)
y_values=x_values**2
# print(y)
plt.figure(figsize=(5,5))
plt.plot(x_values,y_values,label="squared")
plt.legend()
plt.plot(x_values,x_values)
```

```python
x=np.linspace(0,10*np.pi,1000)
y=np.sin(x)
plt.figure(figsize=(6,3),layout="constrained")
plt.plot(x,y)
plt.title("Quad")
```

```python
x=np.random.uniform(0,1,100)
y=np.random.uniform(0,1,100)
plt.scatter(x,y)
```

```python
x=np.random.uniform(0,1,(10,10))

plt.imshow(x)
```

```python
fig,axs=plt.subplots(2,2,figsize=(8,8),layout='constrained')
axs[0,0].imshow(x)
axs[0,1].plot(x_values,y_values)
axs[0,1].plot(x_values,x_values)
```
